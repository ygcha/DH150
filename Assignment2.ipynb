### Step 4: Do regression with linear regression
Using the feature you've identified in x and the target variable y, perform linear regression with sklearn's LinearRegression.
Be sure to use a train/test split

```python
import numpy as np
import seaborn as sns
import sklearn.linear_model
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression

# Select a linear regression model
model = sklearn.linear_model.LinearRegression()
```

```python
x = df2[['Age']]
y = df2['Target']

df2.plot.scatter(x='Age', y='Target', title='Scatterplot of Age and Target');
```

```python
print(df2.corr())
print(df2.describe())
```

```python
X = df2['Age'].values.reshape(-1, 1)
y = df2['Target'].values.reshape(-1, 1)
```

```python
# Train, Test, Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

# Coefficient 
print(model.coef_, model.intercept_)
```

```python
# Make a prediction
X_sample = X_test
y_pred = model.predict(X_test)
for i in range(len(y_test)):
    print(y_test[i], y_pred[i])
```

```python
# Visualize what the predictions are for this model

plt.plot(X_test,y_test,'ko')

X_new = np.linspace(-0.15,0.15)
X_new = X_new.reshape(-1,1)
y_pred = model.predict(X_new)
plt.plot(X_new, y_pred)

plt.show()
```

```python
# Visualize what the predictions are for this model

plt.plot(X_train,y_train,'bo')

X_new = np.linspace(-0.15,0.15)
X_new = X_new.reshape(-1,1)
y_pred = model.intercept_ + model.coef_ * X_new
plt.plot(X_new, y_pred)

plt.show()
```

```python
# Visualize what the predictions are for this model

plt.plot(X_train,y_train,'bo')
plt.plot(X_test,y_test,'ko')

X_new = np.linspace(-0.15,0.15)
X_new = X_new.reshape(-1,1)
y_pred = model.intercept_ + model.coef_ * X_new
plt.plot(X_new, y_pred)

plt.show()
```

```python
y_pred = model.predict(X_test)
df_preds = pd.DataFrame({'Actual': y_test.squeeze(), 'Predicted': y_pred.squeeze()})
print(df_preds)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print('MSE = ', mean_squared_error(y_test, model.predict(X_test)))
print('R^2 = ', r2_score(y_test, model.predict(X_test)))
```
